# **Print 함수 - 특별한 문자**
- \n : 새로운 줄
- \t : 탭(보통 4칸 띄움)
- \\' : 작은 따옴표(') 출력
- \\" : 큰 따옴표(") 출력
- \\\ : 슬래시(\\) 출력
- end 파라미터 : 끝 문자 변경
- sep 파라미터 : 구분 문자 변경

```python
print("파이썬\t반가워\n우리 친해지자\n\n또 만나")
```

```python
print("Hello World! Hello Python!")
```

```python
# 끝 문자 변경 : end 파라미터

print("Hello")
print('Powerful Python')
print("Fun Python")

# 한줄로 인쇄
```

```python
# 구분 문자 변경 : sep 파라미터

# 1칸 공백으로 구분
print("Hello","Easy", "Python")

# 탭으로 구분


# 쉼표(,)로 구분

```

* input() 함수
  - 키보드로 입력된 값을 문자열(string, 문자들이 모인 것)로 반환
  - input() 함수의 인수는 사용자 입력을 돕기 위한 안내 문구 등을 표시하는 문자열 작성

```python
string = input("인사말을 입력하세요>")
```

```python
print(string)
```

```python
number = input("숫자를 입력하세요>")
print(number)
```

# **문자열 함수**

* input() 함수의 입력 자료형은 항상 문자열이므로 입력받은 문자열을 숫자 연산에 활용하기 위해 숫자로 변환
* int() 함수
  - 문자열을 int 자료형(-1, 0, 1, 2, 3)으로 변환
* float() 함수 
  - 문자열을 float 자료형(3.14, 0.12)으로 변환
* str() 함수
  - 숫자를 문자열로 변환
* len() 
  - 문자열 길이 구하는 함수

#관계 연산자
: 두 개의 피연산자를 비교하는데 사용, 결과는 True(1)/Fasle(0)

- x==y : x와 y가 같은가?
- x!=y : x와 y가 다른가?
- x>y : x가 y보다 큰가?
- x<y : x가 y보다 작은가?
- x>=y : x가 y보다 크거나 같은가?
- x<=y : x가 y보다 작거나 같은가?

```python
radius = 100
flag = (radius > 32)
print(flag)
print(int(flag))
```
